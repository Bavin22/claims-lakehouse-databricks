```python
from pyspark.sql import functions as F
from src.claims_project.config import TABLES
```

```python
silver_claims = spark.table(TABLES['silver_claims'])
display(silver_claims.limit(5))
```

```python
silver_providers = silver_claims.select("ProviderID", "ProviderSpecialty","ProviderLocation")\
    .dropDuplicates(["ProviderID"])\
    .withColumn("ProviderId", F.trim(F.col("ProviderID")))\
    .withColumn("ProviderSpecialty", F.upper(F.col("ProviderSpecialty")))\
    .withColumn("ProviderLocation", F.upper(F.col("ProviderLocation")))\
    .filter(F.col("ProviderID").isNotNull())
display(silver_providers.limit(5))
```
