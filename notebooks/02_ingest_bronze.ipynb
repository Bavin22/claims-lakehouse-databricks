```python
from pyspark.sql.functions import (current_timestamp, input_file_name, lit)
```

```python
from src.claims_project.config import (RAW_PATH,TABLES)
```

```python
#Reading dataset

claims_raw = spark.read.option("header", True).csv(f"{RAW_PATH}/claims.csv")
display(claims_raw.limit(5))
```

```python
claims_bronze = claims_raw.withColumn("ingestion_timestamp", current_timestamp())
display(claims_bronze.limit(5))
```

```python
claims_bronze.write.format("delta").mode("overwrite").option("mergeshchema", "true").saveAsTable(TABLES["bronze_claims"])
```

```python
bronze_claims_df = spark.table("bronze_claims")
display(bronze_claims_df.limit(5))
```
